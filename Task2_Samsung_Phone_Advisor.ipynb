**Create Samsung Phone Dataset**

```python
import pandas as pd

data = [
    ["Galaxy S23 Ultra", "2023", "6.8 AMOLED", "5000mAh", "200MP", "12GB", "256GB", 1199],
    ["Galaxy S22 Ultra", "2022", "6.8 AMOLED", "5000mAh", "108MP", "12GB", "256GB", 999],
    ["Galaxy S21 FE", "2021", "6.4 AMOLED", "4500mAh", "12MP", "8GB", "128GB", 699],
    ["Galaxy A54", "2023", "6.4 AMOLED", "5000mAh", "50MP", "8GB", "128GB", 449]
]

columns = ["model", "release", "display", "battery", "camera", "ram", "storage", "price"]
df = pd.DataFrame(data, columns=columns)
df
```

**RAG Module (Retrieval)**

```python
def retrieve_phone(model_name):
    return df[df["model"].str.contains(model_name, case=False)]
```

**Multi-Agent System**

```python
##Agent 1 – (Data Extractor)
def data_extractor(phone_df):
    if phone_df.empty:
        return None
    row = phone_df.iloc[0]
    return row.to_dict()

##Agent 2 – Review Generator
def review_generator(phone1, phone2=None):
    if phone2:
        return (
            f"{phone1['model']} has a better camera ({phone1['camera']}) "
            f"and newer performance compared to {phone2['model']}."
        )
    return f"{phone1['model']} is a balanced phone with strong features."

```

**Unified Question Answer System**

```python
def ask_question(question):
    question = question.lower()

    if "compare" in question:
        models = question.replace("compare", "").split("and")
        p1 = data_extractor(retrieve_phone(models[0].strip()))
        p2 = data_extractor(retrieve_phone(models[1].strip()))
        return review_generator(p1, p2)

    phone = data_extractor(retrieve_phone(question))
    return phone
```

**Example Queries**

```python
ask_question("Compare Galaxy S23 Ultra and S22 Ultra")
```

```python
ask_question("Galaxy S23 Ultra")
```
